```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
climate = pd.read_csv('Adaptation_Innovation_in_Water_Sector_in_Africa.csv')
climate.head()
```

```python

from matplotlib.ticker import FuncFormatter

# Create a figure and axis object
fig, ax = plt.subplots()

# Plot the line graph
ax.plot(climate['Year'], climate['Water_Stress_Index'], color='blue', marker='o', linestyle='-')

# Adding title and labels for the line graph
ax.set_title('Water Stress Index and Water Adaptation Technologies over Time')
ax.set_xlabel('Year')
ax.set_ylabel('Water Stress Index')


# Set x-axis limits
ax.set_xlim(1999, 2017)  # Adjust the limits as needed

# Format x-axis tick labels to remove decimals
ax.xaxis.set_major_formatter(FuncFormatter(lambda x, _: '{:.0f}'.format(x)))


# Twin the x-axis to create a secondary y-axis for the bar graph
ax2 = ax.twinx()

# Plot the bar graph
ax2.bar(climate['Year'], climate['Water_related_Adaptation_technologies'], color='green', alpha=0.5)

# Adding labels for the bar graph
ax2.set_ylabel('Water Adaptation Technologies (#)')


# Displaying plot
plt.show()
```
